```python
import pandas as pd

df = pd.read_csv('Customer Preference Survey (Responses) - Form Responses 1.csv')
```

```python
df.head(5)
```

```python
df=df.drop([0])
```

```python
df=df.fillna(0)
df=df.drop(['Timestamp','playstation5'],axis=1)
```

```python
df.iloc[1].unique()
```

```python
df=df.replace('ไม่เคยซื้อ',0)
df=df.replace('ไม่เคย',0)
df=df.replace('ไม่',0)
df=df.replace('ไม่เคยซือ',0)
df=df.replace('เคยซื้อ',1)
df=df.replace('เคย',1)
```

```python
missing = df.isnull().sum()
print(missing)
```

```python
hot_encoded_df=df
```

```python
hot_encoded_df.head(5)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
hot_encoded_df = hot_encoded_df.applymap(encode_units)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(hot_encoded_df, min_support=0.1, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```python
rules[ (rules['lift'] >= 1) &
       (rules['confidence'] >= 0.5) ].sort_values(by='lift',ascending=False)
```

```python
rules['count1']=rules["antecedents"].apply(lambda x: len(x))
rules['count2']=rules["consequents"].apply(lambda x: len(x))
```

```python
rules[ (rules['lift'] >= 1) &
      (rules['count1']==1)&
      (rules['count2']==1)&
       (rules['confidence'] >= 0.5) ].sort_values(by='lift',ascending=False).head(20)
```

```python
item_item_matrix = pd.DataFrame(index=hot_encoded_df.columns,columns=hot_encoded_df.columns)
item_item_matrix
```

```python
from scipy.spatial.distance import cosine

for i in range(0,len(item_item_matrix.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(item_item_matrix.columns)) :
      # Fill in placeholder with cosine similarities
      item_item_matrix.iloc[i,j] = 1 -cosine(hot_encoded_df.iloc[:,i],hot_encoded_df.iloc[:,j])
```

```python
links = item_item_matrix.rename_axis('related item', 
                                     axis='columns').stack().reset_index()
links.columns = ['item', 'related item','value']
```

```python
links_filtered=links.loc[ (links['value'] > 0.1) & 
                         (links['item'] != links['related item']) ]
```
